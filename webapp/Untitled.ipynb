```python
import pyaudio
import wave
import speech_recognition as sr

def record_audio(output_file, duration=10, chunk=1024, channels=1, rate=44100):
    audio = pyaudio.PyAudio()

    # Open stream
    stream = audio.open(format=pyaudio.paInt16,
                        channels=channels,
                        rate=rate,
                        input=True,
                        frames_per_buffer=chunk)

    frames = []

    # Record audio
    for i in range(0, int(rate / chunk * duration)):
        data = stream.read(chunk)
        frames.append(data)

    # Stop Recording
    stream.stop_stream()
    stream.close()
    audio.terminate()

    # Save audio to file
    with wave.open(output_file, 'wb') as wf:
        wf.setnchannels(channels)
        wf.setsampwidth(audio.get_sample_size(pyaudio.paInt16))
        wf.setframerate(rate)
        wf.writeframes(b''.join(frames))

def segment_audio(input_file):
    recognizer = sr.Recognizer()

    # Load audio file
    with sr.AudioFile(input_file) as source:
        audio_data = recognizer.record(source)

    # Perform speaker diarization
    try:
        transcription = recognizer.recognize_google(audio_data)
        print("Transcription: " + transcription)
        
        # Simple rule-based segmentation
        segments = []
        current_speaker = None
        for line in transcription.split('\n'):
            if line.startswith("Person 1:"):
                if current_speaker != 1:
                    segments.append([])
                    current_speaker = 1
                segments[-1].append(line[len("Person 1:"):])
            elif line.startswith("Person 2:"):
                if current_speaker != 2:
                    segments.append([])
                    current_speaker = 2
                segments[-1].append(line[len("Person 2:"):])
        
        # Print segments
        for i, segment in enumerate(segments, start=1):
            print(f"Person {i}:", ' '.join(segment))
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand the audio.")
    except sr.RequestError as e:
        print("Could not request results from Google Speech Recognition service; {0}".format(e))

if __name__ == "__main__":
    output_file = "conversation.wav"
    record_audio(output_file)
    segment_audio(output_file)
```

```python
import pyaudio
import wave

def record_audio(output_file, duration=10, chunk=1024, channels=1, rate=44100):
    audio = pyaudio.PyAudio()

    # Open stream
    stream = audio.open(format=pyaudio.paInt16,
                        channels=channels,
                        rate=rate,
                        input=True,
                        frames_per_buffer=chunk)

    frames = []

    # Record audio
    for i in range(0, int(rate / chunk * duration)):
        data = stream.read(chunk)
        frames.append(data)

    # Stop Recording
    stream.stop_stream()
    stream.close()
    audio.terminate()

    # Save audio to file
    with wave.open(output_file, 'wb') as wf:
        wf.setnchannels(channels)
        wf.setsampwidth(audio.get_sample_size(pyaudio.paInt16))
        wf.setframerate(rate)
        wf.writeframes(b''.join(frames))

def segment_audio(input_file):
    audio = wave.open(input_file, 'rb')
    rate = audio.getframerate()
    channels = audio.getnchannels()
    chunk = 1024
    frames = []
    current_speaker = None
    segment_counter = 1

    while True:
        data = audio.readframes(chunk)
        if not data:
            break
        frames.append(data)
        if len(frames) == int(rate / chunk):
            audio_segment = b''.join(frames)
            # Perform speaker recognition here if needed
            speaker_id = 1 if current_speaker == 2 else 2
            output_filename = f"speaker_{speaker_id}_segment_{segment_counter}.wav"
            with wave.open(output_filename, 'wb') as wf:
                wf.setnchannels(channels)
                wf.setsampwidth(audio.getsampwidth())
                wf.setframerate(rate)
                wf.writeframes(audio_segment)
            frames = []
            current_speaker = speaker_id
            segment_counter += 1

    audio.close()

if __name__ == "__main__":
    output_file = "conversation.wav"
    record_audio(output_file)
    segment_audio(output_file)
```
